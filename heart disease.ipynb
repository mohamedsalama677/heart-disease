##  The Features :
### 1 - First Feature (age) Refer To ---> Age In Years .
### 2 - Second Feature (sex) Refer To ---> (1 = Male, 0 = Female) .
### 3 - Third Feature (cp) Refer To   ---> Chest Pain Type . ( If You Want to Know about cp : https://archive.ics.uci.edu/ml/datasets/heart+disease) 
### 4 - Fourth Feature (trestbps) Refer To ---> Resting Blood Pressure (If You Want to Know about trestbps : https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3598145/ ) .
### 5 - Fifth Feature (chol) Refer To ---> Serum Cholestoral In mg/dl  .
### 6 - Sixth Feature (fbs) Refer To ---> (Fasting Blood Sugar - 120 mg/dl) (1 = True, 0 = False) .
### 7 - Seventh Feature (restecg) Refer To ---> Resting Electrocardiographic Results( If You Want to Know about restecg : https://www.nhs.uk/conditions/electrocardiogram/) .
### 8 - Eightth Feature (thalach) Refers To --> Maximum Heart Rate Achieved . 
### 9 - Ninth Feature (exang) Refers To --> Exercise Induced Angina (1 = yes, 0 = no) .
### 10 - Tenth Feature (oldpeak) Refers To -->ST Depression Induced by Exercise Relative To Rest . (If You Want to Know About ST Depression : https://en.wikipedia.org/wiki/ST_depression ) 
### 11 - eleventh Feature (slope) Refers To --> ST segment / Heart Rate Slope(If You Want To Know About : https://pubmed.ncbi.nlm.nih.gov/3739881/) .
### 12 - twelveth Feature (ca) Refers To --> Abbreviation For Several Diseases  .
### 13 - Thirteenth Feature (thal) Refers To -->thalassemia is a blood disorder passed down through families (inherited) in which the body makes an abnormal form or inadequate amount of hemoglobin  .
### 14 - Fourteenth Feature (target) Refers To --> Heart patient or not .

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X
```

```python
 # Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # take an object from StandardScaler 
X_train = sc.fit_transform(X_train) # convert X_train 
X_test = sc.transform(X_test) # convert X_test 
```

## Making The Model 
### We will use the best model of this algorthims 


## First We will try a Logistic Regression

```python
# Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier_LG = LogisticRegression(random_state = 0)
classifier_LG.fit(X_train, y_train) # fit the model in train data

# Predicting the Test set results
y_pred_LG = classifier_LG.predict(X_test)
```

#### Evaluate The Algorthim  

```python
cmf_LG = metrics.confusion_matrix(y_test,y_pred_LG)
print(cmf_LG)

#graph of confusion matrix
sns.set(font_scale=1.4) # for label size
sns.heatmap(cmf_LG, annot=True, annot_kws={"size": 16}) # font size
plt.show()


# calculate the Accuracy ,precision and recall
print("Accuracy",metrics.accuracy_score(y_test, y_pred_LG))
print("precision = ", metrics.precision_score(y_test,y_pred_LG))
print("recall = ", metrics.recall_score(y_test,y_pred_LG))
```

## Second We will try a Naive Bayes

```python
# Training the Naive Bayes model on the Training set
from sklearn.naive_bayes import GaussianNB
classifier_NV = GaussianNB()
classifier_NV.fit(X_train, y_train # fit the model in the train data

# Predicting the Test set results
y_pred_NV = classifier_NV.predict(X_test)

```

#### Evaluate The Algorthim  

```python
# Making the Confusion Matrix
cmf_NV = metrics.confusion_matrix(y_test, y_pred_NV)
sns.set(font_scale=1.4) # for label size
sns.heatmap(cmf_NV, annot=True, annot_kws={"size": 16}) # font size
plt.show()


# calculate the Accuracy ,precision and recall
print("Accuracy",metrics.accuracy_score(y_test, y_pred_NV))
print("precision = ", metrics.precision_score(y_test,y_pred_NV))
print("recall = ", metrics.recall_score(y_test,y_pred_NV))
```

## Third We will try a  Decision Tree
